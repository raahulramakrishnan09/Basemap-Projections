```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
```

```python
fig=plt.figure(figsize=(8,8))
m =Basemap()
m.drawcoastlines()
```

```python
m=Basemap(projection='ortho',lat_0=0, lon_0=0)
m.drawmapboundary(fill_color='aqua')
m.fillcontinents(color='coral',lake_color='aqua')
m.drawcoastlines()
plt.show()
```

```python
fig=plt.figure(figsize=(8,8))
m=Basemap(projection='ortho',lat_0=50,lon_0=0)
m.drawmapboundary(fill_color='aqua')
m.fillcontinents(color='coral',lake_color='aqua')
m.drawcoastlines()
plt.show()
```

```python
fig=plt.figure(figsize=(8,6),edgecolor='w')
m=Basemap(projection='cyl',llcrnrlat=-30,urcrnrlat=30,
          llcrnrlon=-80,urcrnrlon=80)
m.drawcoastlines()
```

```python
fig=plt.figure(figsize=(8,6),edgecolor='w')
m=Basemap(projection='cyl',llcrnrlat=-30,urcrnrlat=30,
          llcrnrlon=-80,urcrnrlon=80)
m.shadedrelief(scale=2.0)
m.drawparallels(np.linspace(-90,90, 13))
m.drawmeridians(np.linspace(-180,180,30))
```

```python
fig=plt.figure(figsize=(8,6),edgecolor='w')
m=Basemap(projection='lcc',resolution=None,width=8E6,height=5E6,lat_0=15,lon_0=73)
m.etopo(scale=0.5,alpha=0.5)
x,y=m(79,11.8)
plt.plot(x,y,'ok',markersize=3)
plt.text(x,y,' Tamil Nadu',fontsize=8);
```
